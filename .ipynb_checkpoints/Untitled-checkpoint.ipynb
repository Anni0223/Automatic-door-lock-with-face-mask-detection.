# 3. Converting data and labels list into array

```python
import numpy as np
```

```python
data = np.array(data,dtype="float32")
```

```python
labels = np.array(labels)
```

# 4. Creating train and test split.

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(data,labels,test_size=.20,stratify = labels,random_state=42)
```

# 5. Image Data Generator.

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
aug = ImageDataGenerator(
    rotation_range=20,
    zoom_range=0.15,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.15,
    horizontal_flip=True,
    fill_mode="nearest")
```

```python
#Image Augmentation is a way of applying different types of transformation techniques on actual
# images, thus producing copies of the same image with alterations. This helps to train deep learning
#models on more image variations than what is present in the actual dataset.
```

# 6. Training

```python
# These values are after trial and error.
```

```python
Learning_rate = 1e-4  # ALways give LR less, loss is calculated properly and we get accuracy soon.
```

```python
Epochs = 20
```

```python
BS = 32
```

```python
from tensorflow.keras.applications import MobileNetV2
```

```python
from tensorflow.keras.layers import Input
```

```python
base_model = MobileNetV2(weights = 'imagenet',include_top = False,input_tensor = Input(shape=(224,224,3)))
```

### We have not included the top layer, we will add a top layer manually.

# 6.b Creating the top layer. 

```python
from tensorflow.keras.layers import AveragePooling2D
from tensorflow.keras.layers import Dropout
from tensorflow.keras.layers import Flatten
from tensorflow.keras.layers import Dense
```

```python
headModel = base_model.output
headModel = AveragePooling2D(pool_size=(7, 7))(headModel)
headModel = Flatten(name="flatten")(headModel)
headModel = Dense(128, activation="relu")(headModel)
headModel = Dropout(0.5)(headModel)
headModel = Dense(2, activation="softmax")(headModel)
```

```python
from tensorflow.keras.models import Model
```

```python
model = Model(inputs=base_model.input, outputs=headModel)
```

```python
# loop over all layers in the base model and freeze them so they will
# *not* be updated during the first training process
```

```python
for layer in base_model.layers:
	layer.trainable = False
```

# 6.c Compiling the model.

```python
# compile our model
```

```python
from tensorflow.keras.optimizers import Adam
```

```python
optim = Adam(lr=Learning_rate, decay=Learning_rate/ Epochs)
model.compile(loss="binary_crossentropy", optimizer=optim,metrics=["accuracy"])
```

```python
H = model.fit(
    aug.flow(X_train, y_train, batch_size=BS),
    steps_per_epoch=len(X_train) // BS,
    validation_data=(X_test,y_test),
    validation_steps=len(X_test) // BS,
    epochs=Epochs)
```

# 1. Predictions.

```python
predictions = model.predict(X_test, batch_size=BS)
```

```python
predictions
```

```python
predictions = np.argmax(predictions, axis=1)
```

```python
predictions
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test.argmax(axis=1), predictions,target_names=lb.classes_))
```

# Saving Model

```python
model.save("mask_detector.model", save_format="h5")
```

# Training data visualization

```python
import matplotlib.pyplot as plt
```

```python
N = Epochs
plt.style.use("ggplot")
plt.figure()
plt.plot(np.arange(0, N), H.history["loss"], label="train_loss")
plt.plot(np.arange(0, N), H.history["val_loss"], label="val_loss")
plt.plot(np.arange(0, N), H.history["accuracy"], label="train_acc")
plt.plot(np.arange(0, N), H.history["val_accuracy"], label="val_acc")
plt.title("Training Loss and Accuracy")
plt.xlabel("Epoch #")
plt.ylabel("Loss/Accuracy")
plt.legend(loc="lower left")
plt.savefig("plot.png")
```
